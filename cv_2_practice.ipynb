```python
import cv2
import numpy as np

# Read the original image
original_image = cv2.imread('171.jpg')

# Define rotation range
angles = range(0, 360, 1)

# Output directory
output_directory = 'C:/Users/hp/OneDrive/Pictures/Rotated Pictures/'

# Iterate through angles
for angle in angles:
    # Make a copy of the original image
    rotated_img = np.array(original_image)
    rows, cols, shp = rotated_img.shape
    rotation_matrix = cv2.getRotationMatrix2D((cols / 2, rows / 2), angle, 1.0)
    rotated_img = cv2.warpAffine(rotated_img, rotation_matrix, (cols, rows))

    # Save the rotated image with a filename indicating the angle
    output_filename = f'{output_directory}rotated_image_{angle}.jpg'
    cv2.imwrite(output_filename, rotated_img)

print('Rotation completed.')
```
