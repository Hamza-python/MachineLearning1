```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = 'jaguar-pace.jpg'
img = cv2.imread(path)
plt.imshow(img)
plt.show()
```

```python
img1 = cv2.rectangle(img, pt1 = (70,100), pt2 = (410,300), color = (255,0,0), thickness = 3)
plt.imshow(img1)
plt.show()
```
